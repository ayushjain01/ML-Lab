#### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### Loading dataset

```python
dataset = pd.read_csv(r'./Datasets/Advertising.csv')
dataset.head()
```

### Visualizing Given Data

```python
graphSheet = plt.figure(figsize=(15,10))
graphSheet.add_subplot(3,3,1)
plt.scatter(dataset['TV'],dataset['sales'], c ="orange")
plt.xlabel("Money Spent on TV")
plt.ylabel("sales")
graphSheet.add_subplot(3,3,2)
plt.scatter(dataset['radio'],dataset['sales'],c='pink')
plt.xlabel("Money Spent on Radio")
plt.ylabel("sales")
graphSheet.add_subplot(3,3,3)
plt.scatter(dataset['newspaper'],dataset['sales'],c='teal')
plt.xlabel("Money Spent on Newspaper")
plt.ylabel("sales")
```

#### Splitting the dataset into training and testing set for TV VS Sales

```python
X_train,x_test,y_train,y_test=train_test_split(dataset["TV"],dataset["sales"],test_size=0.3)
X_train = X_train.to_numpy().reshape(-1,1)
x_test = x_test.to_numpy().reshape(-1,1)
y_train = y_train.to_numpy().reshape(-1,1)
y_test = y_test.to_numpy().reshape(-1,1)
```

### Performing Linear Regression for money spent on TV vs Sales

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

#### Predicting

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

#### Visualising

```python
plt.scatter(X_train,y_train,color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color ='blue')
plt.title('TV vs Sales (Training set)')
plt.xlabel('Money Spent on TV')
plt.ylabel('Sales')
plt.show()
plt.scatter(x_test, y_test, color = 'yellow')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('TV vs Sales (Training set)')
plt.xlabel('Money Spent on TV')
plt.ylabel('Sales')
plt.show()
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=(mse)**(1/2)
rsq = r2_score(y_test,y_pred)
print("MSE : ", mse)
print("RSQ : ",rsq)
```
